```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv(r'G:\CYBER SECURITY ALL MODULES\3rd SEMESTER\Dataset\Social_Network_ads.csv')
```

```python
dataset.head()
```

```python
## Using iloc function we have seperated the columns
```

```python
len(dataset)
```

```python
x=dataset.iloc[:,[1,2,3]].values
```

```python
x
```

```python
y=dataset.iloc[:,-1].values
```

```python
y
```

```python
## For converting 0 & 1 we have used labelencoder function
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
x[:,0]=le.fit_transform(x[:,0])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_test)
```

```python
len(x_train)
```

```python
## From Sklearn function we have imported Naive Bayes Algoritham
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)  # purchased 1  not purchased 0
```

```python
x_test  # 1 male   0 Female
```

```python
## Accuracy Test Score
```

```python
model.score(x_test,y_test)
```
